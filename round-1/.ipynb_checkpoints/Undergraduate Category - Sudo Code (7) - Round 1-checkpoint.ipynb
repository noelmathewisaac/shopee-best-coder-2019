```python
## Reading in categories
categories_df = pd.read_csv('Extra Material 2 - keyword list_with substring.csv')
```

```python
## Preparing categories into a dictionary
categories_df.head()
categories = categories_df['Keywords'].tolist()
```

```python
categories

## Split each string into a list
cleaned = []
for cat in categories:
    to_append = cat.split(', ')
    cleaned.append(to_append)

## Map each list to their group no. in a dictionary
final_categories = {}
for i in range(len(cleaned)):
    final_categories[i] = cleaned[i]
```

```python
## Finding out the longest category name - no longer used
longest_string = 0
for k, v in final_categories.items():
    for item in v:
        curr = len(item.split(' '))
        if len(item.split(' ')) >longest_string:
            longest_string = curr
print(longest_string)
```

```python
import nltk
import re
import itertools

## imported itertools originally but ended up unused
```

```python
## Filter for words and spaces - mistake here: need to keep numbers and '-'
pattern = re.compile('([^\s\w]|_)+')
def clean_names(title):
    title = pattern.sub('', title)
    title = title.lower()
    title = title.strip()
    return title
```

```python
## Apply cleaning function to df
keyword_df['cleaned_step_1'] = keyword_df['name'].apply(clean_names)
```

```python
keyword_df.head()
```

```python
## Function to identify matched group numbers
def checker(entry):
    matched_words = {} ## To keep track of all words matched
    to_delete = [] 
    for key, value in final_categories.items():
        for item in value:
            if item in entry:
                to_append = True
                for k in matched_words:
                    if k==item:
                        to_append = False ## We keep the smaller index, which was already found
                    elif k in item:
                        to_delete.append(k) ## If there's a substring already matched, delete it later
                    elif item in k:
                        to_append = False ## Else if we are currently processing a substring, throw it away
                if to_append:
                    matched_words[item] = key ## Add it if it passes the above checks
    result = []
    for item in to_delete:
        if item in matched_words:
            del matched_words[item] ## Remove substrings
    for key, value in matched_words.items():
        if not value in result:
            result.append(value) ## Reduce into a list
    result.sort() 
    return result 
```

```python
## Check the df
keyword_df["final"] = keyword_df["cleaned_step_1"].apply(checker)
```

```python
## Export the results
new_df = keyword_df[["index", "final"]]
new_df.columns=['index','groups_found']
new_df.to_csv("solution.csv", index=False)
```
